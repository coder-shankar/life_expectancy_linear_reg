```python
# let's replace the life expectancy of missing value with mean
df = df.rename(columns = {'Life expectancy ': 'Life expectancy'})

df['Life expectancy'].mean()
```

```python
 df['Life expectancy'] = df['Life expectancy'].fillna(value = df['Life expectancy'].mean())
```

```python
df['Life expectancy'].isnull().sum()
```

```python
#how values are related
plt.figure(figsize = (20,15))
sns.heatmap(df.corr(),annot = True)
```

```python
# Life expectency and schooing is highly correleted
plt.figure(figsize = (15,10))
sns.scatterplot(data=df,y='Life expectancy',x='Schooling')
```

```python
plt.figure(figsize = (15,10))
sns.scatterplot(data=df,y='Life expectancy',x='Adult Mortality')
```

```python
#Though its bad practices to apply mean for missing data if data is large
df['Adult Mortality'] = df['Adult Mortality'].fillna(value = df['Adult Mortality'].mean())
df['Income composition of resources'] = df['Income composition of resources'].fillna(value = df['Income composition of resources'].mean())
    
```

```python
data_plot = df
data_plot = data_plot.drop(['Status','Country'], axis=1)
```

```python
sns.set_style("darkgrid")

graph = sns.jointplot(y="Schooling", x="Life expectancy", data=data_plot,
                  kind="reg", truncate=False,
                  color="#0000FF", height=7)
```

```python
# As Schooing increase life expectency is also increasing
```

```python
#Lets drop null values for now
df = df.dropna()
df.isnull().sum()
```

```python
y = df['Life expectancy']
X = df.drop(['Status','Country', 'Life expectancy'], axis=1)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression().fit(X_train, y_train)
```

```python

reg.score(X_test,y_test)
```
